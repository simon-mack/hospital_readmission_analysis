```python

enc_test = encounters_by_patient.loc[encounters_by_patient['Patient ID'] == 'DS1|10451'].sort_values(by = 'Encounter End Date', ascending = True).reset_index(drop=True)

enc_test
```

```python
#Add variables to keep track of counts

index = 0
readm_count=0
for enc_date in enc_test['Encounter Start Date']:
    if index == 0:
        init_number = index+1
        init_enc_date = enc_date
        index_offset= 1
        print(f"Initial Encounter {init_number}: {init_enc_date}")
        dis_date = enc_test['Encounter End Date'].loc[0]
        print(f"Initial Discharge: {dis_date}")
        index+=1
        print(f"index: {index}\n")
    else:
        if (enc_date-dis_date).days > 30:
            print(f"Encounter {index+1}: {enc_date}\nMore than 30 days after\nInitial Discharge: {dis_date} ")
            print(f"Therefore Encounter {index+1}({enc_date}) is counted as an 'Initial encounter'")
            init_enc_date = enc_date
            init_number = init_number+1
            print(f"New initial encounter {init_number}: {init_enc_date}")
            dis_date = enc_test['Encounter End Date'].loc[index]
            print(f"New initial Discharge: {dis_date}")
            index_offset= index+1
            print(f"index: {index+1}\n")
            
        else:
            print(f"Encounter {index+1}: {enc_date} \nLess than 30 days after\nInitial Discharge: {dis_date}")
            if index-index_offset==0:
                print(f"Therefore initial encounter {init_number}({init_enc_date}) is counted as a new 'under 30 day readmission'")
                readm_count+=1
                print(f"index: {index+1}\n")
            else:
                print(f"Therefore initial encounter {init_number}({init_enc_date}) is counted as an 'under 30 day readmission' but NOT added to the count")
                print(f"index: {index+1}\n")
        index +=1
    print(f"Running total of Initial Encounters: {init_number}")
    print(f"Running total of Initial Encounters with readmission: {readm_count}\n")
print(f"Total number of Initial Encounters: {init_number}")        
print(f"Total of Initial Encounters with readmission: {readm_count}\n")
```

```python
#Loop through and create temporary dataframe of of encounters for each patient ID
total_init_enc=0
total_readmissions=0
for patient_id in encounters_by_patient['Patient ID'].unique():
    enc_temp = encounters_by_patient.loc[encounters_by_patient['Patient ID'] == patient_id].sort_values(by = 'Encounter End Date', ascending = True).reset_index(drop=True)
    index = 0
    readm_count=0
    for enc_date in enc_temp['Encounter Start Date']:
        if index == 0:
            init_number = index+1
            init_enc_date = enc_date
            index_offset= 1
            dis_date = enc_temp['Encounter End Date'].loc[0]
            index+=1
        else:
            if (enc_date-dis_date).days > 30:
                init_enc_date = enc_date
                init_number = init_number+1
                dis_date = enc_temp['Encounter End Date'].loc[index]
                index_offset= index+1   
            else:
                if index-index_offset==0:
                    readm_count+=1
            index +=1
        
    #print(f"Total number of Initial Encounters for patient {patient_id}: {init_number}")        
    #print(f"Total of Initial Encounters with readmission for patient {patient_id}: {readm_count}\n")
    total_init_enc = total_init_enc+init_number
    total_readmissions = total_readmissions+readm_count
    #print(f"Running Total number of Initial Encounters: {total_init_enc}")        
    #print(f"Running Total of Initial Encounters with under 30 days readmission : {total_readmissions}\n")
    
print(f"Total number of Initial Encounters: {total_init_enc}")        
print(f"Total of Initial Encounters with under 30 days readmission : {total_readmissions}\n")
```
